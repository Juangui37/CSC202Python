this is showing you every word in the song with the word count for each word

```python
from textblob import TextBlob
blob = TextBlob(song)
blob.word_counts
```

This code is removing the stop words in the english language

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
stopwords.add("n't")
stopwords
```

This is print all the words in the song with the stop words removed

```python
words = [ (w, f) for w,f in blob.word_counts.items() if w not in stopwords]
words
```

This is sorting all the words in the song with the stop words removed

```python
from operator import itemgetter
words = sorted(words, key = itemgetter(1), reverse=True)
words
```

this creates a dataframe with the word and frequency using the pandas package

```python
import pandas as pd

# create a data frame from a list of tuples (each element will be a column)
df = pd.DataFrame(words, columns = ['word', 'frequency'])
df[:20]
```

this creates a word cloud of all the most used words - the stopwords in the song

```python
from wordcloud import WordCloud
wordcloud = WordCloud().generate(song)
wordcloud.to_image()
```

## Now we're going to tackle n-grams (2b)

Repeat the above using n-grams. An n-gram corresponds to a set of n consecutive words. A bigram is an n-gram where n = 2 and a trigram is an n- gram where n = 3. For example, the sentence how are you? has bigrams of
     
how are and are you. Using textblob, n-grams can be accessed using blob.ngrams(n). Generate a table and/or wordcloud showing the most common bigrams and/or trigrams.

```python
from textblob import TextBlob
blob = TextBlob(song)
b2 = blob.ngrams(2)
b2
```

```python
ngramslist = [  (e[0].lower() + ' ' + e[1].lower()) for e in b2         ]
```

```python
ngramslist
```

```python
from collections import Counter
```

```python
counterngrams = Counter(ngramslist)
```

```python
wordtuple = [   (w, f) for w, f in counterngrams.items() if w not in stopwords      ]
```

```python
orderedwords = sorted(wordtuple, key = itemgetter(1), reverse=True)
orderedwords
```

```python
import pandas as pd

df = pd.DataFrame(orderedwords, columns = ['word', 'frequency'])
df[:20]
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud().generate_from_frequencies(counterngrams)
wordcloud.to_image()
```

### Now I'm going to print each sentence and there polarity

```python
polaritytuple = [   (str(w), w.sentiment.polarity) for w in blob.sentences if w not in stopwords      ]
polaritytuple
```

### Now, I'm going to print the sentences in order of highest positive polarity 

```python
from operator import itemgetter
pt = sorted(polaritytuple, key = itemgetter(1), reverse=True)
pt
```

### To finish, I'm going to display a dataframe of the top 5 positive and negative polarity sentences

```python
df = pd.DataFrame(pt, columns = ['sentence', 'polarity'])
df[:5]
```
