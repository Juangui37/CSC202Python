````python
 with open('beemovie.txt', 'r') as f : # open a file for reading
        text = f.read() # stores the text of the file in text
````

````python
from textblob import TextBlob
blob = TextBlob(text)
blob
````

````python
import nltk
nltk.download('punkt')
````

````python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
  
stopwords = set(stopwords.words('english'))
newstopwords = stopwords.union("n't")
newstopwords

word_tokens = word_tokenize(text)
 
filtered_sentence = [w for w in word_tokens if not w.lower() in newstopwords]
 
filtered_sentence = []
 
for w in word_tokens:
    if w not in newstopwords:
        filtered_sentence.append(w)
 
print(word_tokens)
print(filtered_sentence)
````

````python
words = [ (w, f) for w,f in blob.word_counts.items() if w not in stopwords]
words
````

````python
# sort the list
from operator import itemgetter

# sort the word list by frequency (stored in index 1 of each tuple) in reverse order (highest to lowest)
words = sorted(words, key = itemgetter(1), reverse=True)
words
````

````python
import pandas as pd

# create a data frame from a list of tuples (each element will be a column)
df = pd.DataFrame(words, columns = ['word', 'frequency'])
df[:20]
````
